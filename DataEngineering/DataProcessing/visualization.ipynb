# 1. Graph

## 그래프의 종류
- 도수분포표(히스토그램) : 단일 집단의 분포를 확인
- 선형도(Line Plot) : 시간에 따른 추세(경향) 확인
- 산점도(Scatter Plot) : 두 변수간의 관계를 파악
- Bar Chart : 변수 간 차이를 파악
- Pie Chart : 카테고리 별 비중의 차이 파악
- Box Plot : 집단별 특정 변수의 분포 차이 파악, 분포를 나타내기엔 히스토그램이 가장 좋은 방법이지만 여러개의 분포를 동시에 비교할 때에는 Box Plot으로 대략적인 차이 파악

## (1) Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

### 그래프 겹쳐 그리기

```python
x = np.arange(0, 10, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.plot(x, np.cos(x))
# 한 셀에 그래프 함수를 여러개 쓰면 겹쳐서 그릴 수 있음.
```

### 선의 색깔, 스타일 바꾸기

```python
plt.plot(x,np.sin(x), color='blue', linestyle='--')
plt.plot(x, np.cos(x), color='red', linestyle=':')
```

### 축의 경계 바꾸기
특정 그래프에만 해당하는 것이 아닌 전체에 적용되는 설정  
plt의 plot에서 설정하는 것이 아니라 plt에서 xlim/ylim을 바로 불러옴

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11) # x축의 범위 조정
plt.ylim(-1.5, 1.5) # y축의 범위 조정
```

### x축과 y축의 척도를 같게 조절

```python
plt.plot(x, y, color='green', linestyle='--')
plt.plot(x, np.cos(x), color='red', linestyle=':')
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.axis('equal') # x, y 의 척도가 같아짐(1의 간격이 동일)
```

### 그래프의 레이블 붙이기

```python
plt.plot(x, y, color='green', linestyle='--')
plt.plot(x, np.cos(x), color='red', linestyle=':')
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.axis('equal')
plt.title('sin and cos wave')
plt.xlabel('x') # x축 레이블
plt.ylabel('sin(x)') # y축 레이블
```

### 범례 달기

```python
plt.plot(x, y, color='green', linestyle='--', label='sin')
plt.plot(x, np.cos(x), color='red', linestyle=':', label='cos')
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.axis('equal')
plt.title('sin and cos wave')
plt.xlabel('x')
plt.ylabel('y')
plt.legend() # 각 곡선의 범례 표시
```

### 그래프의 크기 조정하기

```python
plt.figure(figsize=(8, 6), dpi=200) # 그래프를 그리기 전에 크기 조정
plt.plot(x, y, color='green', linestyle='--', label='sin')
plt.plot(x, np.cos(x), color='red', linestyle=':', label='cos')
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.axis('equal')
plt.title('sin and cos wave')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

## (3) 다중그래프
하나의 그림에 여러개의 그래프를 붙여서 동시에 보여주고 싶을 때

### 방법 1. subplot
이 방법의 단점은 다중그래프 안의 개별적인 그래프를 튜닝할 수 없다는 것이다.  
그래프에 공통적으로 사용되는 성질을 지정하거나 그래프의 수가 너무 많아지는 경우 한계가 있다.  
따라서 주로 정석적인 방법 2로 다중그래프를 나타낸다.

```python
x = np.arange(0, 10, 0.1)
np.cos(x)
```

```python
plt.figure(figsize=(8, 7))

# 첫번째 그래프
plt.subplot(1, 3, 1) # 1행 3열 중 첫번째임을 의미
plt.plot(x, np.sin(x), color='red', label='sin(x)')
plt.title('sin curve')
plt.legend()

# 두번째 그래프
plt.subplot(1, 3, 2)
plt.plot(x, np.cos(x), color='green')
plt.title('cos curve')

# 세번째 그래프
plt.subplot(1, 3, 3)
plt.scatter(x, np.sin(x))
plt.title('scatter sin curve')
```

### 방법 2. fig, axes, subplots

```python
fig, axes = plt.subplots(1, 3, figsize = (8, 4)) # fig 전체 그림, axes 개별 그래프
axes[0].plot(x, np.sin(x), c='red', label='sin(x)')
axes[0].legend()
axes[1].plot(x, np.cos(x), c='green', label='cos(x)')
axes[1].legend()
axes[2].scatter(x, np.sin(x), label='scatter sin(x)')
axes[2].legend()
```

### 축 공유

```python
fig, axes = plt.subplots(1, 3, figsize=(8, 4), sharey=True) # 축 공유 : sharex / sharey
axes[0].plot(x, np.sin(x), c='red', label='sin(x)')
axes[0].legend()
axes[1].plot(x, np.cos(x), c='green', label='cos(x)')
axes[1].legend()
axes[2].scatter(x, np.sin(x), label='scatter sin(x)')
axes[2].legend()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(4, 8), sharex=True) # 축 공유 : sharex / sharey
axes[0].plot(x, np.sin(x), c='red', label='sin(x)')
axes[0].legend()
axes[1].plot(x, np.cos(x), c='green', label='cos(x)')
axes[1].legend()
axes[2].scatter(x, np.sin(x)+2, label='scatter sin(x)')
axes[2].legend()
```

### axes의 알고리즘적으로 접근
axes를 2차원 행렬로 접근할 수 있음  
for 반복문 등을 적용하여 자동으로 그래프를 채울 수 있음

```python
fig, axes = plt.subplots(3, 3, figsize=(8, 8), sharex=True, sharey=True)
axes[0, 0].plot(x, np.sin(x))
axes[1, 1].plot(x, np.cos(x))
axes[2, 2].plot(x, x+2)
```

```python
fig, axes = plt.subplots(3, 3, figsize=(8, 8), sharex=True, sharey=True)
for i in range(3):
  for j in range(3):
    if (i+j)%2==0:
      axes[i, j].plot(x, np.sin(x), c='red')
    else :
      axes[i, j].plot(x, np.cos(x), c='green')
```

다수의 히스토그램 그리기

```python
fig, axes = plt.subplots(2, 2, sharex=True, figsize=(8, 6))
for i in range(len(fig.axes)):
  sample_size=10**(i+1) #데이터의 갯수 : 10, 100, 1000, 10000
  sample = np.random.normal(size=sample_size) # 표준정규분포에서 데이터 추출
  axes[i//2, i%2].hist(sample, bins=100)
  axes[i//2, i%2].set_title(str(sample_size))
```

### style 설정, 소수점 조정

```python
plt.plot(x, np.cos(x))
style1={'size':10, 'color':'green'}
style2={'size':10, 'color':'red'}

# 소수점 설정하기
# '{:.2f}' : 소수 둘째자리까지 나타냄, format(나타낼 숫자)
# 딕셔너리 앞에 **을 붙여서 텍스트 안에 명령어로 넣어주면 코드를 알아서 해석
plt.text(10, np.cos(10), '{:.2f}'.format(np.cos(10)), **style1)
plt.text(15, np.cos(15), '{:.3f}'.format(np.cos(15)), **style2)

# 텍스트의 위치 조정하기
#plt.text(10+0.2, np.cos(10), '{:.2f}'.format(np.cos(10)), **style1)
#plt.text(15+0.2, np.cos(15), '{:.3f}'.format(np.cos(15)), **style2)
```

### scatter plot을 이용한 자동화
### f를 이용한 formating

```python
x = np.arange(0, 20, 0.1)
style = {'size':10, 'color':'black'}
delta = 0.05
plt.figure(figsize=(8,6))
plt.plot(x, np.cos(x))
for i in np.arange(0, 20, 2): # 0에서부터 2씩 증가하면서 10개의 점 찍기
  plt.scatter(i, np.cos(i), color='red')
  plt.text(i+delta, np.cos(i)+delta, f'{i}, {np.cos(i):.2f}', **style) # 중괄호 안의 연산을 수헹 후 문자열 타입으로 변환해줌  /  :.2f 로 동시에 소수점 조정 -  위 문법과 비교
```

### 텍스트를 화살표로 강조하기

```python
x = np.arange(0, 20, 0.1)
style ={'size':10, 'color':'black'}

arrowstyle = {'arrowstyle':'->'}

plt.figure(figsize=(8, 6))
plt.plot(x, np.cos(x))
plt.text(10, np.cos(10), '10', **style)
plt.text(15, np.cos(15), '15', **style)

# ammotate : 텍스트('maximum')에서부터 xy로 가는 화살표 그리기
plt.annotate('maximum', xy=(6.28, 1), xytext=(8, 0.9), arrowprops = arrowstyle) # xy=(6.28, 1) : 가르킬 위치, xytext=(8, 0.9) :텍스트의 위치
plt.scatter(6.28, 1, c='red')
```

### datetime 타입 변환 -  split과 map

```python
# split과 final을 시간형 데이터로 바꿔주기 
from datetime import timedelta
```

```python
def convert_time(s):
  h, m, s = map(int, s.split(':')) # map(int, s.split(':')) - s를 : 기준으로 split하고. int형으로 바꾼 후 h, m, s에 하나씩 맵핑
  return timedelta(hours=h, minutes=m, seconds=s).total_seconds()
```

```python
convert_time("04:35:43")
```

```python
df['split_sec'] = df['split'].apply(convert_time)
df['final_sec'] = df['final'].apply(convert_time)
df
```

```python
with sns.axes_style('white'):
  sns.jointplot('split_sec', 'final_sec', df, kind='hex') # 의미있는 결과를 얻을 수는 없음
```

```python
df['split_frac'] = 1 - 2*(df['split_sec']/df['final_sec']) 
# 음수(negative pace) : 처음보다 나중 속도가 빠름, 양수(positive pace) : 처음에  비해 나중 속도가 느림)
df
```

### axvline 경계선 그리기
