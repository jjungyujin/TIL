## 데이터 Rescaling

: 서로 다른 단위와 범위를 가진 데이터를 같은 범위를 갖도록 통일
- Standard Scaling
: 데이터를 구성하는 각 열이 평균 0, 표준편차 1을 갖도록 변환  
표준화를 통해 표준정규분포를 따르도록 함
- Min-Max Scaling
: 데이터를 구성하는 각 열이 최소값이 0, 최대값이 1을 갖도록 변환

```python
# Standard Scaling
(df2-df2.mean())/df2.std()
```

```python
df2
```

```python
# MIn-Max Scaling
# 최솟값 min, 최댓값 max이면 값의 범위의 크기는 max-min
# (특정 값)k - min : max - min = x : 1
(df2-df2.min())/(df2.max()-df2.min())
```

## Grouping 후 Aggregation 연산
리스트이름.agg(연산자 함수명)  
1. 그루핑
2. 함수를 적용(그룹별로 대표하는 하나의 값 구함)
3. 그룹별 대표값을 모아 combine

```python
df.groupby('Team')['Salary'].agg(np.mean)
df.groupby('Team')['Age'].agg(np.mean)
df.groupby('Position')['Salary'].agg(np.mean)
```

```python
# 여러개의 연산 한번에 하기
df.groupby('Position')['Salary'].agg([np.min, np.max, np.mean, np.median])
```

```python
# 사용자 정의 연산과 함께
# 입력 : 여러개의 값(리스트)
# 출력 : 하나의 요약된 값
def myFunc(x : list):
  return (np.max(x)-np.min(x))

df.groupby('Position')['Salary'].agg(myFunc)
```

## Grouping 후 Transform 연산
1. 그룹핑
2. 함수를 적용(그룹별 대표값 구하기)
3. **원데이터에 브로드캐스팅** : Aggregation과의 차이  
원데이터와 행의 수가 같게 나오므로 원데이터에 새로운 열로 추가할 수 있음

```python
df['TeamAvgSalary'] = df.groupby('Team')['Salary'].transform(np.mean)
```

```python
df
```

```python
# 유형 1
# 입력 : 다수(리스트)
# 출력 : 단일
# transform --> 브로드캐스팅 (팀별로 같은 값을 갖게 됨)
def firstFunc (x : list):
  return (np.max(x)-np.min(x))

df['type1'] = df.groupby('Team')['Salary'].transform(firstFunc)
```

```python
df
```

```python
# 유형 2
# 입력 : 다수(리스트)
# 출력 : 다수(리스트)
# transform --> 브로드캐스팅 아님 (같은 팀이더라도 서로 다른 값을 갖게 됨)
def secFunc (x : list):
  return x-np.mean(x)

df['type2'] = df.groupby('Team')['Salary'].transform(secFunc)
```

```python
df
```

## Grouping 후 Filter 연산
1. 그룹핑
2. 함수를 적용(그룹별 대표값 구하기) 및 **필터링**
3. 원데이터에 브로드캐스팅 

```python
# 그룹 후 필터링 사용자 정의함수
# 입력 : 다수(리스트)
# 출력 : True or False
def myFilter(x):
  return x['Salary'].mean() > 5000000

df.groupby('Team').filter(myFilter)
```

```python
# 예제
def myCounter(x):
  return len(x)<15

# team을 기준으로 split된 그룹이 차례대로 x에 들어감
# True에 해당하는 그룹들로 combine
df.groupby('Team').filter(myCounter)
```

## 데이터 합치기 방법 - how 옵션 이용하기
- Inner Join  
: 두 데이터프레임에 공통적으로 포함되는 데이터만 포함
- Left Join  
: 왼쪽 데이터 기준으로 결합
- Right Join  
: 오른쪽 데이터 기준으로 결합
- Outer Join  
: 두 데이터프레임 중 하나라도 존재할 경우 포함

```python
df_customer = pd.DataFrame(
    {
    'id' : [1, 2, 3, 4],
     'name' : ['Tom', 'Jenny', 'James', 'Dan']
    }
)

df_info = pd.DataFrame(
    {
    'id' : [2, 3, 4, 5],
    'age' : [31, 20, 40, 70],
    'sex' : ['F', 'M', 'M', 'F']
    }
)
```

```python
pd.merge(df_customer, df_info, how='inner', on='id')
```

```python
pd.merge(df_customer, df_info, how='left', on='id')
# 왼쪽을 기준으로 합칠 때 중복되지 않는 행에 대한 새로운 열의 값은 NaN이 된다
```

```python
pd.merge(df_customer, df_info, how='right', on='id')
```

```python
pd.merge(df_customer, df_info, how='outer', on='id')
```

## Validate 활용

```python
df_customer = pd.DataFrame(
    {
    'id' : [1, 2, 3, 4],
     'name' : ['Tom', 'Jenny', 'James', 'Dan']
    }
)

df_info = pd.DataFrame(
    {
    'id' : [2, 2, 3, 4, 5],
    'age' : [31, 21, 20, 40, 70],
    'sex' : ['F', 'F', 'M', 'M', 'F']
    }
)
```

```python
pd.merge(df_customer, df_info, on='id')
# 같은 id를 가지는 서로 다른 행이 2개 존재함
# id는 한 명에게 대응되어야 하는데 df_info의 중복된 값 때문에 문제 발생
```

```python
# 현재 1대 1 조건을 만족하지 않으므로 오류 발생(오류가 무조건 나쁜 게 아님)
# validate를 사용하면 내가 원하는 조건에 충족하는지 확인할 수 있음
pd.merge(df_customer, df_info, on='id', validate='1:1')
```

```python
import pandas as pd

android_devices = pd.read_csv('/content/drive/MyDrive/android_devices.csv')
user_device = pd.read_csv('/content/drive/MyDrive/user_device.csv')
user_usage = pd.read_csv('/content/drive/MyDrive/user_usage.csv')
```

```python
android_devices
# 브랜드의 디바이스(모델의 상위그룹명), 모델명 정보
```

```python
user_device
# 사용자별 디바이스 정보
```

```python
user_usage
# 사용자별 사용량 정보
```

```python
# 데이터프레임 분석 목적 : 브랜드별로 스마트폰의 사용시간 분석하기
# 1. user_usage와 user_device 합치기 : 사용자별 디바이스, 사용량 정보 하나로 담기
df2 = pd.merge(user_usage, user_device, how='inner', on='use_id')
```

```python
df2
```

```python
# 2.android_devices와 df2 합치기 : 사용자별 디바이스에 해당하는 브랜드 찾기
pd.merge(df2, android_devices, left_on='device', right_on='Model')
# 데이터프레임의 행이 늘어난 이유 : 중복 데이터가 존재함
# 사용자가 여러 디바이스를 사용할 수는 있으나
# android_devices 데이터프레임의 Model은 고유한 값을 가져야 함
```

```python
android_devices['Model']
```

```python
len(android_devices['Model'].unique())
```

```python
# 약 2000개 중복 -> 중복 제거 필요
android_devices.drop_duplicates(subset = 'Model', inplace=True)
android_devices
```

```python
result = pd.merge(df2, android_devices, left_on='device', right_on='Model')
result
```
